# customer dataset
    This table contains the details of the customer

```python
%%sql 
--preview 
select top 10 * from customers_dataset
```

```python

%%sql
--how many unique customers 
select count(distinct customer_unique_id) as NO_of_customers
from customers_dataset
```

```python
%%sql 
--statewise citywise customers list

select  top 100 customer_state,
customer_city,
count(distinct customer_unique_id) as no_of_customers
from customers_dataset 
where customer_state <>'' and customer_city <> ''
group by customer_state,customer_city
order by customer_state,customer_city
```

# Customer dataset + Order dataset:
        This dataset gives information on both the details of the customer and their order details. 


```python
%%sql 
--preview
select  top 10 * 
from customers_dataset cd
inner join orders_dataset od
on cd.customer_id=od.customer_id
```

```python
%%sql
--valuable_customers
select tab3.* 
from (
    select dense_rank() over(order by tab2.purchase_count desc) as rank,
    tab2.*
    from (
        select tab1.customer_unique_id ,
        tab1.customer_state,
        tab1.customer_city,
        count(*) as purchase_count 
        from (
            select cd.customer_state,
            cd.customer_city,
            cd.customer_unique_id,
            od.order_status
            from customers_dataset cd
            inner join orders_dataset od
            on cd.customer_id=od.customer_id) tab1
        where tab1.customer_unique_id <> '' and tab1.order_status <> 'cancelled' and tab1.order_status <> 'unavailable'
        group by tab1.customer_state,tab1.customer_city,tab1.customer_unique_id) tab2 
    ) tab3

 where rank<=3

   
```

```python
%%sql
--order status wise customer count
select tab1.order_status,
count(distinct tab1.order_id) as order_count 
from (
    select cd.customer_unique_id,
    od.order_status,
    od.order_id
    from customers_dataset cd
    inner join orders_dataset od
    on cd.customer_id=od.customer_id
    where cd.customer_unique_id <>'') tab1
where tab1.order_status <>''
group by tab1.order_status

   
```

# Order dataset
    This table contains the details of the order_id,status and time of purchase.

```python
%%sql
--preview
select top 10 * from orders_dataset
```

```python
%%sql 
--average time taken for delivery

select concat(cast(avg(tab1.day_difference) as varchar(10)),' Days') as day_average 
from(
    select  
    datediff(day,cast(substring(order_approved_at,1,10) as date),cast(substring(order_delivered_customer_date,1,10) as date)) as day_difference
    from orders_dataset
    where order_approved_at <> '' and order_delivered_customer_date <> '') tab1
```

```python
%%sql
--daywise purchase count

select top 10
tab1.day,
count(*) as purchase_count
from(
    select 
    substring(order_purchase_timestamp,1,10) as day
    from orders_dataset) tab1
group by tab1.day
order by purchase_count desc
 
```

# order_items_dataset

```python
%%sql
--preview
select top 10 * 
from order_items_dataset
```

```python
%%sql 
--product quantity in a order
select order_id,
count(*) as product_quantity
from  order_items_dataset
group by order_id
having count(*) > 10
order by product_quantity desc
```

# order_items_dataset + product_dataset
    This dataset gives details of the purchased Products.

```python
%%sql 
--preview
select top 10 *
from order_items_dataset oid
inner join products_dataset pd
on oid.product_id=pd.product_id
```

```python
%%sql 
--preview

select top 10 tab2.seller_id,
trans.product_category_name_english as product_category_name,
tab2.product_count
from (
    select seller_id,
    product_category_name,
    count(distinct product_id) as product_count
    from(
        select
        oid.seller_id,
        pd.product_category_name,
        oid.product_id
        from order_items_dataset oid
        inner join products_dataset pd
        on oid.product_id=pd.product_id) tab1
    where seller_id <> '' and product_category_name <>''
    group by seller_id,product_category_name) tab2
inner join product_category_name_translation trans
on tab2.product_category_name=trans.product_category_name
order by tab2.product_count desc
```

# Joining more tables:

```python
%%sql 
--preview
select top 10 *
from (
    select tab1.customer_id,
    tab1.product_id,
    cd.customer_state,
    cd.customer_city
    from(
        select od.customer_id,
        oid.product_id
        from order_items_dataset oid
        inner join orders_dataset od
        on oid.order_id=od.order_id) tab1
    inner join customers_dataset cd
    on tab1.customer_id=cd.customer_id) tab2
inner join products_dataset pd
on tab2.product_id=pd.product_id
```

```python
%%sql
--statewise citywise product category count

select top 100
tab4.customer_state,
tab4.customer_city,
tab4.product_category_name,
count(*) as purchase_count
from(
    select tab3.customer_state,
    tab3.customer_city,
    trans.product_category_name_english as product_category_name
    from(
        select tab2.customer_state,
        tab2.customer_city,
        pd.product_category_name
        from (
            select tab1.customer_id,
            tab1.product_id,
            cd.customer_state,
            cd.customer_city
            from(
                select od.customer_id,
                oid.product_id
                from order_items_dataset oid
                inner join orders_dataset od
                on oid.order_id=od.order_id) tab1
            inner join customers_dataset cd
            on tab1.customer_id=cd.customer_id) tab2
        inner join products_dataset pd
        on tab2.product_id=pd.product_id) tab3 
    inner join product_category_name_translation trans
    on tab3.product_category_name=trans.product_category_name) tab4
where tab4.customer_state <> '' and tab4.customer_city <> '' and tab4.product_category_name <> ''
group by tab4.customer_state,tab4.customer_city,tab4.product_category_name
--order by purchase_count desc
```

# Pivoting the above column(Product Category Name)

```python
%%sql
--pivoting the data

select tab5.customer_state,
sum(case when tab5.pc='bed_bath_table' then tab5.purchase_count else NULL end) as bed_bath_table,
sum(case when tab5.pc='health_beauty' then tab5.purchase_count else NULL end) as health_beauty,
sum(case when tab5.pc='sports_leisure' then tab5.purchase_count else NULL end) as sports_leisure,
sum(case when tab5.pc='housewares' then tab5.purchase_count else NULL end ) as housewares,
sum(case when tab5.pc='furniture_decor' then tab5.purchase_count else NULL end) as furniture_decor

from(
    select 
    tab4.customer_state,
    tab4.customer_city,
    tab4.pc,
    count(*) as purchase_count
    from(
        select tab3.customer_state,
        tab3.customer_city,
        trans.product_category_name_english as pc
        from(
            select tab2.customer_state,
            tab2.customer_city,
            pd.product_category_name
            from (
                select tab1.customer_id,
                tab1.product_id,
                cd.customer_state,
                cd.customer_city
                from(
                    select od.customer_id,
                    oid.product_id
                    from order_items_dataset oid
                    inner join orders_dataset od
                    on oid.order_id=od.order_id) tab1
                inner join customers_dataset cd
                on tab1.customer_id=cd.customer_id) tab2
            inner join products_dataset pd
            on tab2.product_id=pd.product_id) tab3 
        inner join product_category_name_translation trans
        on tab3.product_category_name=trans.product_category_name) tab4
    where tab4.customer_state <> '' and tab4.customer_city <> '' and tab4.pc <> ''
    group by tab4.customer_state,tab4.customer_city,tab4.pc) tab5
group by customer_state
order by 1
```
